```python
 Write a matcher called 'Swimming' that finds both occurrences of the phrase "swimming vigorously" in the text.
```

```python
from spacy.matcher import Matcher
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
with  open("owlcreek.txt", "r",encoding="utf8") as txt:
    text = txt.read()

doc = nlp(text)

matcher = Matcher(doc.vocab)
```

```python
pattern = [{"LOWER": "Swimming"}, {"LOWER":"swimming vigorously"}]
```

pattern = [{"LOWER": "Swimming"},
          {"LOWER":"swimming vigorously"}]



```python
matcher.add("Swimming", [pattern])
```

```python
match = matcher(doc)
```

```python
for matcher_id, start, end in match:
    span = doc[start:end]
    print(span.text)
```
